```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
import sympy as sp

i = np.linspace(-10,10,400)
x=sp.symbols('x')
y=-x
print("x(dot)=")
display(y)
v=sp.integrate(-y,x)


print('v(x)=')
display(v)
xdot=[]
V=[]
for X in i:
    xdot.append(y.evalf(subs={x:X}))
    V.append(v.evalf(subs={x:X}))
Table=pd.DataFrame()
print(type(xdot[0]))
Table['x']=i
Table['dx/dt']=np.array(xdot).astype(np.float64)
Table['v(x)']=np.array(V).astype(np.float64)
display(Table)
a=np.empty(len(Table))
a.fill(0)
fig=go.Figure()
fig.add_trace(go.Scatter(x=Table['x'],y=Table['dx/dt'],mode='lines',name=sp.latex(y)))
fig.add_trace(go.Scatter(x=Table['x'],y=Table['v(x)'],mode='lines',name='$'+sp.latex(v)+'$'))


fig.show()
```
